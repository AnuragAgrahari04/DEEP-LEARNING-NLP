```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
#preprocess the data
#drop irrelevant data

data = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
data
```

```python
# Encode categorical variable

label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
onehot_encoder_geo = OneHotEncoder()
geo_encoded = onehot_encoder_geo.fit_transform(data[['Geography']])
geo_encoded_df = pd.DataFrame(geo_encoded.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
#combining columns with the original data
data = pd.concat([data.drop('Geography',axis = 1),geo_encoded_df],axis=1)
```

```python
data
```

```python
#Divide the dataset into features and targets

X = data.drop('EstimatedSalary', axis=1)
Y = data['EstimatedSalary']
```

```python
#split the data into training and testing sets

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

#scale the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# save the encoders and scaler

with open('label_encoder_gender.pkl', 'wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('onehot_encoder_geo.pkl', 'wb') as file:
    pickle.dump(onehot_encoder_geo,file)

with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler,file)
```

### ANN Regression Problem Statement

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
## Build ANN Model

model  = Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),  ## HL! Connected with input layer 1
    Dense(32,activation='relu'),  ## HL2
    Dense(1)  ## Output Layer   ## Default Activation function is linear activation function for regression
]
)
```

```python
#compile the model
model.compile(optimizer='adam',loss='mean_absolute_error',metrics=['mae'])
```

```python
model.summary()
```

```python
# setup the tensorboard

log_dir = "regressionLogs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
# setup early stopping

early_stopping = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
# Train the model

history = model.fit(
    X_train, Y_train, validation_split=0.2, epochs=100,
    batch_size=32, callbacks=[early_stopping,tensorflow_callback]
)
```

```python
# Evaluate the model on the test data

test_loss, test_mae = model.evaluate(X_test, Y_test)
print('Test MAE:', test_mae)
```
